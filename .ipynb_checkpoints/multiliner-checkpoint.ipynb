```python
import pandas as pd
import pandas as pd
from sklearn.preprocessing import LabelEncoder
file3=r'‪C:\Users\keert\Downloads\\50_Startups.csv'.replace('\u202a','')
department=pd.read_csv(file3)
department
label_encoder = LabelEncoder()
department['State'] = label_encoder.fit_transform(department['State'])
print("\nDataFrame after Label Encoding:")
print(department)
```

```python
x=department.iloc[:,:-1].values
x
```

```python
y=department.iloc[:,1].values
y
```

```python
from sklearn.linear_model import LinearRegression
regressor1=LinearRegression()
regressor1.fit(x,y)
```

```python
print(regressor1.coef_)
print(regressor1.intercept_)
```

```python
import numpy as np
ypredict=np.array([[165349.20 ,136897.80,471784.10,2]])
regressor1.predict(ypredict)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=0)
print(x_train)
print(y_train)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
regressor
```

```python
y_predict=regressor.predict(x_test)
y_predict
```

```python
y_test
```

```python
x_test
```

```python
from sklearn.metrics import r2_score
print('train score',regressor.score(x_train,y_train))
print('test score',regressor.score(x_test,y_test))
print('pred score',r2_score(y_predict,y_test))
```
