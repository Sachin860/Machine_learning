```python
import pandas as pd
df=pd.read_csv(r"A:\sachin drive\framingham_heart_disease.csv")
df1=pd.DataFrame(df)
print(df1.shape)
df1.info()
df1.dropna(how='any',inplace=True)
print(df1.shape)
print(df1)
```

```python
x=df1.iloc[:,:-1].values
print(pd.DataFrame(x))
```

```python
y=df1.iloc[:,15].values
print(pd.DataFrame(y))
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x,y)
```

```python
print(regressor.coef_)
#y=mx+c linear regression formula
print(regressor.intercept_)
```

```python
import numpy as np
ypredict=np.array([[1,39,4,0,0,0,0,0,0,195,106,70,26,80,77]])
ypredict
```

```python
regressor.predict(ypredict)
```

```python
ypredict1=np.array([[0,61,3,1,30,0,0,1,0,225,150,95,28.58,65,103]])
ypredict1
```

```python
regressor.predict(ypredict1) #if >0.5 =1, <0.5=0
```

```python
#logistics linear regresssion
import pandas as pd
#import matplotlib.pyplot as plt
#from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.model_selection import train_test_split
#from sklearn import tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
```

```python
a=df1.drop(columns=['TenYearCHD'])
print(a)
b=df1['TenYearCHD']
print(b)
```

```python
model=LogisticRegression()
model.fit(a,b)
```

```python
y_pred = model.predict(a)
print(y_pred)

print("----------------")
# Print evaluation metrics
print("Accuracy:", accuracy_score(b, y_pred))
print("\nClassification Report:\n", classification_report(b, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(b, y_pred))
```
